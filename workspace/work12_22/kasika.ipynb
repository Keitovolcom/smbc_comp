```python
import pandas as pd
import numpy as np
import plotly as py
import plotly.graph_objects as go
import json
import pandas as pd
```

```python
train = pd.read_csv('C:/Users/wgkdj/smbc_comp/input/train.csv', index_col=0) # 学習用データ
test = pd.read_csv('C:/Users/wgkdj/smbc_comp/input/test.csv', index_col=0)
tree= pd.read_csv('C:/Users/wgkdj/smbc_comp/workspace/work1213/boro_ct_avg.csv')
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
f = open('NTA_map.geojson')
nta_json = json.load(f)
for i in list(range(0,len(nta_json['features']))):
    json_id=nta_json['features'][i]['properties']['ntacode']
    nta_json['features'][i]['id']=json_id
```

```python
data =[]
data.append(go.Choroplethmapbox(geojson=nta_json, 
                                    locations=train.nta.value_counts().index, 
                                    z=train.nta.value_counts().values,
                                    colorbar=dict(title='2015'),
                                    colorscale="YlGn"
                                    ))
data.append(go.Choroplethmapbox(geojson=nta_json, 
                                    locations=test.nta.value_counts().index, 
                                    z=test.nta.value_counts().values,
                                    colorbar=dict(title='2005'),
                                    colorscale="YlGn"
                                    ))
data[0]['visible']=False
data[1]['visible']=True
layout=go.Layout(mapbox_style="carto-positron",
                  mapbox_zoom=9, mapbox_center = {"lat": 40.7, "lon": -73.86})
layout.update(updatemenus=list([
        dict(
            x=-0.05,            y=1,
            yanchor='top',
            buttons=list([
                dict(
                    args=['visible', [True, False]],
                    label='train',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, True]],
                    label='test',
                    method='restyle')]))]))
fig = go.Figure(data=data,layout=layout)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
train['long'] = train['boro_ct'].map(tree.set_index('boro_ct')['boro_ct_avg_long'])
train['lat'] = train['boro_ct'].map(tree.set_index('boro_ct')['boro_ct_avg_lat'])
```

```python
temp_df=train.dropna().loc[train.dropna()['spc_common'].str.contains('oak')][['spc_common','lat','long']]
data=[]
for t in temp_df['spc_common'].value_counts().index:
    data.append(go.Scattermapbox(lat=temp_df.loc[train.dropna()['spc_common']==(t)]['lat'],
                               lon=temp_df.loc[train.dropna()['spc_common']==(t)]['long'],
                               mode='markers',
                                 name=t,
                              marker=go.scattermapbox.Marker(size=5)
                      ))
layout=go.Layout(mapbox_style="carto-positron",
                  mapbox_zoom=9, mapbox_center = {"lat": 40.7, "lon": -73.86})
fig = go.Figure(data=data,layout=layout)
fig.update_layout(margin={"r":0,"t":20,"l":0,"b":0})
fig.show()
```

```python
train_health_0 = train[train['health'] == 0]
train_health_1 = train[train['health'] == 1]
train_health_2 = train[train['health'] == 2]

data = []

# Choropleth map for train_health=0
data.append(go.Choroplethmapbox(
    geojson=nta_json,
    locations=train_health_0.nta.value_counts().index,
    z=train_health_0.nta.value_counts().values,
    colorbar=dict(title='Train Health 0'),
    colorscale="YlGn",
))

# Choropleth map for train_health=1
data.append(go.Choroplethmapbox(
    geojson=nta_json,
    locations=train_health_1.nta.value_counts().index,
    z=train_health_1.nta.value_counts().values,
    colorbar=dict(title='Train Health 1'),
    colorscale="YlGn",
))

# Choropleth map for train_health=2
data.append(go.Choroplethmapbox(
    geojson=nta_json,
    locations=train_health_2.nta.value_counts().index,
    z=train_health_2.nta.value_counts().values,
    colorbar=dict(title='Train Health 2'),
    colorscale="YlGn",
))

# Set the initial visibility
for i in range(1, len(data)):
    data[i]['visible'] = False

layout = go.Layout(
    mapbox_style="carto-positron",
    mapbox_zoom=9,
    mapbox_center={"lat": 40.7, "lon": -73.86}
)

# Update menu for switching between different train_health values
layout.update(updatemenus=list([
    dict(
        x=-0.05,
        y=1,
        yanchor='top',
        buttons=list([
            dict(
                args=[{'visible': [True, False, False]}],
                label='Train Health 0',
                method='restyle'
            ),
            dict(
                args=[{'visible': [False, True, False]}],
                label='Train Health 1',
                method='restyle'
            ),
            dict(
                args=[{'visible': [False, False, True]}],
                label='Train Health 2',
                method='restyle'
            )
        ])
    )
]))

fig = go.Figure(data=data, layout=layout)
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```
