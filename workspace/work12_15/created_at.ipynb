```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
train_date = pd.read_csv('C:/Users/wgkdj/smbc_comp/input/train.csv')
test_date = pd.read_csv('C:/Users/wgkdj/smbc_comp/input/test.csv')
```

```python
train_date['created_at'] = pd.to_datetime(train_date['created_at'])
test_date['created_at'] = pd.to_datetime(test_date['created_at'])
```

```python
date_counts = train_date['created_at'].value_counts().sort_index()

# 
plt.figure(figsize=(15, 5))  # 
sns.lineplot(x=date_counts.index, y=date_counts.values, marker='o', color='skyblue')  # 

# 
plt.title('Number of Records by Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)  # X

# 
plt.show()
```

```python
test_date_counts = test_date['created_at'].value_counts().sort_index()

# 
plt.figure(figsize=(15, 5))  # 
sns.lineplot(x=test_date_counts.index, y=test_date_counts.values, marker='o', color='skyblue')  # 

# 
plt.title('Number of Records by Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)  # X

# 
plt.show()
```

```python
train_date['created_at'] = pd.to_datetime(train_date['created_at'])

# 'weekdays'
train_date['weekdays'] = train_date['created_at'].dt.day_name()

# user_type
grouped_data = train_date.groupby(['user_type', 'created_at']).size().reset_index(name='count')

# 
plt.figure(figsize=(30, 10))

# user_type
sns.lineplot(x='created_at', y='count', hue='user_type', data=grouped_data, marker='o')

# 
plt.title('Number of Records by Date and User Type (Train)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)

# 
plt.legend(title='User Type')
plt.xlim(pd.Timestamp('2015-05-01'), pd.Timestamp('2015-12-31'))
# 
plt.tight_layout()
plt.show()
```
