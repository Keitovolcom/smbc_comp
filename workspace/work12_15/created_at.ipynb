```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
train_date = pd.read_csv('C:/Users/wgkdj/smbc_comp/input/train.csv')
test_date = pd.read_csv('C:/Users/wgkdj/smbc_comp/input/test.csv')
```

```python
train_date['created_at'] = pd.to_datetime(train_date['created_at'])
test_date['created_at'] = pd.to_datetime(test_date['created_at'])
```

```python
date_counts = train_date['created_at'].value_counts().sort_index()

# プロットの設定
plt.figure(figsize=(15, 5))  # グラフのサイズを調整
sns.lineplot(x=date_counts.index, y=date_counts.values, marker='o', color='skyblue')  # 折れ線グラフの作成

# グラフの装飾
plt.title('Number of Records by Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)  # X軸の日付表示を回転させる（任意）

# グラフの表示
plt.show()
```

```python
train_date_counts = train_date['created_at'].value_counts().sort_index()

# testデータの日付ごとのデータの取得と整理
test_date_counts = test_date['created_at'].value_counts().sort_index()

# グラフの設定
plt.figure(figsize=(15, 10))  # グラフのサイズを調整

# Trainデータのプロット
plt.subplot(2, 1, 1)  # 2行1列の1番目のサブプロット
sns.lineplot(x=train_date_counts.index, y=train_date_counts.values, marker='o', color='red')
plt.title('Number of Records by Date (Train)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(20))
# Testデータのプロット
plt.subplot(2, 1, 2)  # 2行1列の2番目のサブプロット
sns.lineplot(x=test_date_counts.index, y=test_date_counts.values, marker='o', color='blue')
plt.title('Number of Records by Date (Test)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)

# グラフの表示
plt.tight_layout()  # レイアウトを調整して重なりを防ぐ
plt.show()
```

```python

# 'created_at'列の日付データがdatetime型であることを確認
train_date['created_at'] = pd.to_datetime(train_date['created_at'])

# 'weekdays'列を作成
train_date['weekdays'] = train_date['created_at'].dt.day_name()

# Trainデータの日付ごとのデータの取得と整理
train_date_counts = train_date['created_at'].value_counts().sort_index()

# Testデータの日付ごとのデータの取得と整理
test_date_counts = test_date['created_at'].value_counts().sort_index()

# グラフの設定
plt.figure(figsize=(15, 10))  # グラフのサイズを調整

# Trainデータのプロット
plt.subplot(2, 1, 1)  # 2行1列の1番目のサブプロット
sns.lineplot(x=train_date_counts.index, y=train_date_counts.values, color='red')
plt.title('Number of Records by Date (Train)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(20))

# Testデータのプロット
plt.subplot(2, 1, 2)  # 2行1列の2番目のサブプロット
sns.lineplot(x=test_date_counts.index, y=test_date_counts.values, color='blue')
plt.title('Number of Records by Date (Test)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)

# グラフの表示
plt.tight_layout()  # レイアウトを調整して重なりを防ぐ
plt.show()
```

```python
train_date['created_at'] = pd.to_datetime(train_date['created_at'])

# 'weekdays'列を作成
train_date['weekdays'] = train_date['created_at'].dt.day_name()

# グループ化して各user_typeの個数を取得
grouped_data = train_date.groupby(['user_type', 'created_at']).size().reset_index(name='count')

# グラフの設定
plt.figure(figsize=(30, 10))

# 各user_typeごとにラインプロット
sns.lineplot(x='created_at', y='count', hue='user_type', data=grouped_data, marker='o')

# グラフの装飾
plt.title('Number of Records by Date and User Type (Train)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)

# 凡例を表示
plt.legend(title='User Type')
plt.xlim(pd.Timestamp('2015-05-01'), pd.Timestamp('2015-12-31'))
# グラフの表示
plt.tight_layout()
plt.show()
```
